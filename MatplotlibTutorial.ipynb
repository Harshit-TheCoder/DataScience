```python
import numpy as np
import matplotlib.pyplot as plt

# Random seed for reproducibility
np.random.seed(42)

# Example data
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.normal(0, 0.2, 100)  # noisy sine wave
categories = ['A', 'B', 'C', 'D']
values = [23, 17, 35, 29]
scatter_x = np.random.rand(50)
scatter_y = np.random.rand(50)
scatter_size = np.random.randint(50, 300, 50)
```

```python
plt.plot(x, y, color='blue', marker='o', linestyle='--')
plt.title("Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
plt.bar(categories, values, color='purple', edgecolor='black')
plt.title("Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
plt.barh(categories, values, color='orange', edgecolor='black')
plt.title("Horizontal Bar Plot")
plt.xlabel("Values")
plt.ylabel("Categories")
plt.show()
```

```python
plt.scatter(scatter_x, scatter_y, s=scatter_size, c='green', alpha=0.7)
plt.title("Scatter Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
plt.scatter(scatter_x, scatter_y, s=scatter_size, c=scatter_size, cmap='viridis', alpha=0.7)
plt.title("Scatter Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
# plt.axis('equal')
plt.title("Pie Chart")
plt.show()
```

```python
days = np.arange(1, 6)
sleep = [7, 8, 6, 11, 7]
eat = [2, 3, 4, 3, 2]
work = [7, 8, 7, 2, 2]
play = [8, 5, 7, 8, 13]

plt.stackplot(days, sleep, eat, work, play, labels=['Sleep', 'Eat', 'Work', 'Play'], colors=['m','c','r','g'])
plt.legend(loc='upper left')
plt.title("Stack Plot")
plt.show()
```
